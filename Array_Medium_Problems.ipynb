# Two Pointer Technique

```python
def twoSum(arr, target):
    # if arr is not sort then sort it first
    left, right = 0, len(arr) - 1
    
    while left < right:
        total = arr[left] + arr[right]
        
        if total == target:
            return True
        elif total < target:
            left += 1
        else:
            right -= 1
    
    return False

arr = [-3, -1, 0, 1, 2]
target = -2
twoSum(arr, target)
```

# Prefix Sum Technique

```python
def preSum(arr):
    
    n = len(arr)
    
    prefixSum = [0] * n
    
    prefixSum[0] = arr[0]
    
    for i in range(1, n):
        prefixSum[i] = prefixSum[i-1] + arr[i]
    
    return prefixSum

arr = [10, 20, 10, 5, 15]
preSum(arr)
```

# Hashing Technique

```python
def isSubset(a, b):
    hash_set = set(a)
    
    for num in b:
        if num not in hash_set:
            return False
        
    return True

a = [11, 1, 13, 21, 3, 7]
b = [11, 3, 7, 1]
isSubset(a,b)
```

# Sliding Window Technique

```python
def maxSum(arr, k):
    
    n = len(arr)
    
    if n <= k:
        print("Invalid")
        return -1
    
    window_sum = sum(arr[:k])
    
    max_sum = window_sum
    
    # Compute the remaining window by removing first element of previous window and adding last element of the current window
    for i in range(n-k):
        window_sum = window_sum - arr[i] + arr[i+k]
        max_sum = max(max_sum, window_sum)
    
    return max_sum


arr = [5, 2, -1, 0, 3]
k = 3
maxSum(arr,k)
```

# Maximum Circular Subarray Sum

```python
def maxCircularSum(arr):
    
    totalSum = 0
    currMaxSum = 0
    currMinSum = 0
    maxSum = arr[0]
    minSum = arr[0]
    
    for i in range(len(arr)):
        
        currMaxSum = max(currMaxSum + arr[i], arr[i])
        maxSum = max(maxSum, currMaxSum)
        
        currMinSum = min(currMinSum + arr[i], arr[i])
        minSum = min(minSum, currMinSum)
        
        totalSum += arr[i]
        
    normalSum = maxSum
    circularSum = totalSum - minSum
    
    if minSum == totalSum:
        return normalSum
    return max(normalSum, circularSum)

arr = [8, -8, 9, -9, 10, -11, 12]
print(maxCircularSum(arr))
```

# Reorder an array according to given index array

```python
def reorderArray(arr, index):
    
    i = 0
    
    while i < len(arr):
        
        if index[i] == i:
            i += 1
            
        else: 
            temp1 = arr[i]
            arr[i] = arr[index[i]]
            arr[index[i]] = temp1
            
            temp2 = index[i]
            index[i] = index[temp2]
            index[temp2] = temp2
    


arr = [10, 11, 12]
index = [1, 0, 2]

reorderArray(arr, index)
print(arr)
```

# Product of Array Except self

```python
def productExceptSelf(arr):
    zeros = 0
    idx = -1
    prod = 1
    
    for i in range(len(arr)):
        if arr[i] == 0:
            zeros += 1
            idx = i
        else:
            prod *= arr[i]
    res = [0] * len(arr)
    
    if zeros == 0:
        for i in range(len(arr)):
            res[i] = prod // arr[i]
            
    elif zeros == 1:
        res[idx] = prod
    
    return res

arr = [10, 3, 5, 6, 2]
output = productExceptSelf(arr)
print(" ".join(map(str, output)))
```

# Find smallest missing number

```python
def findFirstMissing(arr, n, m):
    vec = [0] * m
    
    for i in range(n):
        vec[arr[i]] = 1
        
    for i in range(m):
        if vec[i] == 0:
            return i
        
    return m

arr = [0, 1, 2, 3, 4, 5, 6, 7, 10]
n = len(arr)
m = 11
print("Smallest missing element is", findFirstMissing(arr, n, m))
```

# Count possible triangles

```python
def countTriangles(arr):
    res = 0
    arr.sort()
    
    for i in range(2, len(arr)):
        left, right = 0, i - 1
        
        while left < right:
            
            if arr[left] + arr[right] > arr[i]:
                res += right - left
                
                right -= 1
            else:
                left += 1
    return res

arr = [4, 6, 3, 7]
print(countTriangles(arr))
```

# Subarray with given Sum

```python
def subArraySum(arr, target):
    # initialize the window => sliding window concept
    start, end = 0,0
    
    res = []
    
    curr_total = 0
    
    for i in range(len(arr)):
        
        curr_total += arr[i]
        
        if curr_total >= target:
            
            end = i
            
            if curr_total > target and start < end:
                
                curr_total -= arr[start]
                start += 1
                
            
            if curr_total == target:
                res.append(start + 1)
                res.append(end + 1)
                return res
            
    return [-1] # if no subarray is found

arr = [15, 2, 4, 8, 9, 5, 10, 23]
target = 23
subArraySum(arr, target) # output is [2,5] ie. from position 2 to 5 => 2 + 4 + 8 + 9 = 23
```

# Longest subarray with equal 0s and 1s

```python
def longsubarraylen(arr):
    mp = {}
    
    preSum = 0
    res = 0
    
    for i in range(len(arr)):
        
        preSum += -1 if arr[i] == 0 else 1
        
        if preSum == 0:
            res = i + 1
            
        if preSum in mp:
            res = max(res, i - mp[preSum])
        else:
            mp[preSum] = i
            
    return res

arr = [1, 0, 0, 1, 0, 1, 1]
print(longsubarraylen(arr))
```
