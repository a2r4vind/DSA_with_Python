```python
class MyQueue:
    
    def __init__(self, capacity):
        
        self.capacity = capacity
        
        self.arr = [0] * capacity
        
        self.size = 0
        
        self.front = self.rear = -1
        
    def isEmpty(self):
        response = False
        if self.size == 0:
            self.front = self.rear = -1
            response = True
        return response
    
    def isFull(self):
        return self.size == self.capacity
    
    def __len__(self):
        return self.size
    
    def enqueue(self, data):
        
        # check Queue is full or not
        if self.isFull():
            return "Queue is Full!"
        
        if self.rear == -1:
            self.front = self.rear = 0
        
        self.arr[self.rear] = data
        
        self.rear += 1
        self.size += 1
        
    def dequeue(self):
        
        # check if empty
        if self.isEmpty():
            return "Queue is empty."
        data = self.arr[self.front]
        self.front += 1
        self.size -= 1
        return data
        
    def getFront(self):
        
        # empty check
        if self.isEmpty():
            return "Queue is empty!"
        data = self.arr[self.front]
        return data
    
    def getRear(self):
        
        #empty check
        if self.isEmpty():
            return "Queue is Empty."
        
        data = self.arr[self.rear-1]
        return data
    
    def printQueue(self):
        
        # empty check
        if self.isEmpty():
            return "Queue is empty."
        
        temp = self.front
        
        while temp is not self.rear:
            print(f"{self.arr[temp]}",end="")
            if (temp + 1) is not self.rear:
                print(" -> ",end="")
            temp += 1
        print()
        
        
```
