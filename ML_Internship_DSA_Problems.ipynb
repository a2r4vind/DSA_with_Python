# Two Sum

--> find two numbers in an array that add up to a target

```python
def twoSum(nums, target):
    
    seen = {}
    for i, num in enumerate(nums):
        
        complement = target - num
        
        if complement in seen:
            return [seen[complement], i]
        
        seen[num] = i
    return []
```

# Maximum Subarray ( Kadane's Algorithm)

--> Find the contiguous subarray with the larget sum

```python
def maxSubArray(nums):
    
    max_sum = nums[0]
    curr_sum = nums[0]
    
    for i in range(1, len(nums)):
        curr_sum = max(nums[i], curr_sum + nums[i])
        max_sum = max(max_sum, curr_sum)
    return max_sum
```

# Merge Sorted Arrays

--> Merge two sorted arrays into one sorted array 

```python
def merge(nums1, m, nums2,n):
    
    # start from the end
    i = m - 1 # last element in nums1
    j = n - 1 # last element in nums2
    k = m + n - 1 # last position in merged array
    
    while j >= 0:
        
        if i >= 0 and nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else: 
            nums1[k] = nums2[j]
            j -= 1
        k -= 1
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3
merge(nums1, m, nums2, n)
print(nums1)
```

# Best Time to Buy and Sell Stocks

--> Find maximum profit from stocks prices

```python
def maxProfit(prices):
    
    min_price = float('inf')
    max_profit = 0
    
    for price in prices:
        min_price = min(min_price, price)
        max_profit = max(max_profit, price - min_price)
        
    return max_profit
```

```python
prices = [7,1,5,3,6,4]
maxProfit(prices)
```

# Product of Array Except Self

--> Return array where each element is product of all others

```python
def productExceptSelf(nums):
    
    n = len(nums)
    result = [1] * n
    
    # left products
    left = 1
    for i in range(n):
        result[i] = left
        left *= nums[i]
    
    # right products
    right = 1
    for i in range(n-1,-1,-1):
        result[i] *= right
        right *= nums[i]
        
    return result
```

```python
nums = [1,2,3,4]
print(productExceptSelf(nums))
```

# Longest Substring Without repeating Characters

--> Find length of longest substring without repeating chars

--> Sliding window concept

```python
def lengthOfLongestSubstring(s):
    
    char_set = set()
    left = 0
    max_length = 0
    
    for right in range(len(s)):
        
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
        
    return max_length

s = 'abcabcbb'
print(lengthOfLongestSubstring(s))
```

# Valid Anagram

--> Check if two strings are anagrams


```python
# Hashmap Approach

def isAnagram(s,t):
    
    # basic condition is if lengths are not equal they are not anagram
    if len(s) != len(t):
        return False
    
    count = {}
    
    for char in s:
        count[char] = count.get(char,0) + 1
    
    for char in t:
        if char not in count:
            return False
        
        count[char] -= 1
        if count[char] < 0:
            return False
    return True

s = 'anagram'
t = 'nagaram'
isAnagram(s,t)
```

# Reverse Linked List

--> Reverse a singly linked list

```python
def reverseLL(head):
    
    prev = None
    curr = head
    
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    return prev
```

# Detect Cycle in Linked List 
# ( Floyd's Cycle Detection)

```python
def hasCycle(head):
    if not head:
        return False
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast: # => indicate cycle is present
            return True
    return False
```

--> Implement Queue using Stacks

```python
class MyQueue:
    def __init__(self):
        self.stk1 = [] # push
        self.stk2 = [] # pop
        
    
    def enqueue(self,x):
        self.stk1.append(x)
        
    def dequeue(self):
        self._move()
        return self.stk2.pop()
    
    def peek(self):
        self._move()
        return self.stk2[-1]
    
    def _move(self):
        if not self.stk2:
            while self.stk1:
                self.stk2.append(self.stk1.pop())
```

 # Binary Search

```python
def binarySearch(arr, key):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == key:
            return mid
        elif arr[mid] > key:
            right = mid -1
        else:
            left = mid + 1
            
    return -1

arr = [1,3,5,7,9,11]
key = 7
binarySearch(arr,key)
```

# Maximum Depth of Binary Tree

```python
def maxDepth(root):
    if not root:
        return 0
    
    left_depth = maxDepth(root.left)
    right_depth = maxDepth(root.right)
    
    return max(left_depth, right_depth)
```

# Invert Binary Tree

```python
def invertTree(root):
    
    if not root:
        return None
    
    # swap left and right children
    root.left, root.right = root.right, root.left
    
    # Recursively invert subtrees
    invertTree(root.left)
    invertTree(root.right)
    
    return root
```

# Validate Binary Search Tree

--> Check if tree is valid BST

```python
def isValidBST(root, min_val=float('-inf'), max_val=float('inf')):
    if not root:
        return True
    if root.val <= min_val or root.val >= max_val:
        return False
    
    return (isValidBST(root.left, min_val, root.val) and isValidBST(root.right, root.val, max_val))
```

# Climbing Stairs
 
--> Count ways to climb n stairs( 1 or 2 stairs at a time)

```python
def climbStairs(n):
    if n <= 2:
        return n
    
    prev2 = 1 # ways to each step 1
    prev1 = 2 # ways to reach step 2
    
    for i in range(3, n + 1):
        curr = prev1 + prev2
        prev2 = prev1
        prev1 = curr
    return prev1

climbStairs(5)
```

# House Robber

--> Maximum money robbing houses without adjacent ones

```python
def rob(nums):
    if not nums:
        return 0
    prev2 = 0
    prev1 = 0
    
    for num in nums:
        curr = max(prev1, prev2 + num)
        prev2 = prev1
        prev1 = curr
    
    return prev1

nums = [2,7,9,3,1]
print(rob(nums))
```
