# Two Sum

--> find two numbers in an array that add up to a target

```python
def twoSum(nums, target):
    
    seen = {}
    for i, num in enumerate(nums):
        
        complement = target - num
        
        if complement in seen:
            return [seen[complement], i]
        
        seen[num] = i
    return []
```

```python
nums = [2,7,11,15]
target = 22
twoSum(nums, target)
```

# Maximum Subarray ( Kadane's Algorithm)

--> Find the contiguous subarray with the larget sum

```python
def maxSubArray(nums):
    
    max_sum = nums[0]
    curr_sum = nums[0]
    
    for i in range(1, len(nums)):
        curr_sum = max(nums[i], curr_sum + nums[i])
        max_sum = max(max_sum, curr_sum)
    return max_sum
```

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
print(maxSubArray(nums))
```

# Merge Sorted Arrays

--> Merge two sorted arrays into one sorted array 

```python
def merge(nums1, m, nums2,n):
    
    # start from the end
    i = m - 1 # last element in nums1
    j = n - 1 # last element in nums2
    k = m + n - 1 # last position in merged array
    
    while j >= 0:
        
        if i >= 0 and nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else: 
            nums1[k] = nums2[j]
            j -= 1
        k -= 1
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3
merge(nums1, m, nums2, n)
print(nums1)
```

# Best Time to Buy and Sell Stocks

--> Find maximum profit from stocks prices

```python
def maxProfit(prices):
    
    min_price = float('inf')
    max_profit = 0
    
    for price in prices:
        min_price = min(min_price, price)
        max_profit = max(max_profit, price - min_price)
        
    return max_profit
```

```python
prices = [7,1,5,3,6,4]
maxProfit(prices)
```

# Product of Array Except Self

--> Return array where each element is product of all others

```python
def productExceptSelf(nums):
    
    n = len(nums)
    result = [1] * n
    
    # left products
    left = 1
    for i in range(n):
        result[i] = left
        left *= nums[i]
    
    # right products
    right = 1
    for i in range(n-1,-1,-1):
        result[i] *= right
        right *= nums[i]
        
    return result
```

```python
nums = [1,2,3,4]
print(productExceptSelf(nums))
```

# Valid Parentheses

--> check if string has valid matching parentheses

```python
def isValid(s):
    
    stack = []
    
    mapping = {')':'(','}':'{',']':'['}
    
    for char in s:
        if char in mapping:
            top = stack.pop() if stack else '#'
            if mapping[char] != top:
                return False
        else:
             stack.append(char)
                
    return len(stack) == 0

s = '()[]{}'
print(isValid(s))
```
