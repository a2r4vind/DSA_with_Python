# 1) Check Even or Odd

```python
# Bitwise AND Operation of Number X with 1 will give 1 if number is Odd else will give 0 if numer is 
```

```python
def isEven(n):
    if (n & 1) == 0:
        return True
    else:
        return False
```

```python
n = 13
if isEven(n):
    print("True")
else:
    print("False")
```

# 3) Sum of n Natural Numbers

```python
# Naive approach

def findSum(n):
    sum = 0
    i = 1
    while i <= n:
        sum += i
        i += 1
    return sum
```

```python
n = 5
print(findSum(n))
```

```python
# Recursive approach

def findTotal(n):
    
    if n == 1:
        return 1
    return n + findTotal(n - 1)
```

```python
n = 5
print(findTotal(n))
```

```python
# Expected Approach is using the formula

def findNSum(n):
    
    return n * (n + 1) // 2
```

```python
n = 5 
print(findNSum(n))
```

# 4) Sum of first n Squared numbers


```python
def summation(n):
    sq_sum = (n * (n + 1) * (2 * n + 1)) // 6
    return sq_sum
```

```python
n = 10
print(summation(n))
```

# 5) Swap two numbers

```python
a = 3
b = 8

print(f"Before swapping a and b is {a} and {b} respectively.")

a, b = b, a # inbuilt swapping operation

print(f"After swapping : ")
print("a", a)
print("b",b)
```

```python
# Swapping using temporary variable

a = 3
b = 8

print(f"Before swapping a and b is {a} and {b} respectively.")

temp = a
a = b
b = temp

print(f"After swapping : ")
print("a", a)
print("b",b)
```

```python
# swapping using arithmetic operators

a = 3
b = 8

print(f"Before swapping a and b is {a} and {b} respectively.")

a = a + b
b = a - b # b <= (a + b) - b therefore b = a
a = a - b # a <= (a + b) - (a) therefore a = b

print(f"After swapping : ")
print("a", a)
print("b",b)
```

# 6) Find Closest to n and Divisible by m

```python
def closestNum(n,m):
    rem = abs(n) % m
    result = n - rem
    return result

ip1 = 13
m1 = 4
print(closestNum(ip1,m1))
# output => 12

ip2 = -15
m2 = 6
print(closestNum(ip2, m2))
# output => -18
```

# 7) The Dice Problem

--> The task is to guess the number on the opposite face of the cube

```python
# Since the sum of numbers on the opposite always add up to 7 for example 1 + 6, 2 + 5, 3 + 4

def oppositeFaceOfDice(n):
    if n < 1 or n > 6:
        return f"No face with the number {n} exists in a fair dice."
    else:
        return (7 - n)
```

```python
n1 = 2
print(oppositeFaceOfDice(n1))
n2 = 6
print(oppositeFaceOfDice(n2))
```

# 8) Nth term of AP from first two terms

```python
def nthTermOfAP(a1, a2, n):
    # using the formula of nth term of an AP => T(n) = a1 + (n-1)*(a2-a1)
    return (a1 + (n-1) * (a2-a1))
```

```python
a1 = 2
a2 = 3
n = 4
print(nthTermOfAP(a1, a2, n))
```

# 1) Sum of Digits

```python
def sumOfDigits(n):
    s = str(n)
    sum = 0
    
    for digit in s:
        sum += int(digit)
        
    return sum
```

```python
n = 12345
print(sumOfDigits(n))
```

# 2) Reverse Digits

```python
def reverseDigits(n):
    s = str(n)
    res = ''
    
    for digit in s:
        res = digit + res
    
    return int(res)
```

```python
n = 122
print(reverseDigits(n))
```

# 3) Prime Testing

```python
import math

def isPrime(n): # ==> Time Complexity = O(root(n)) & space complexity = O(1)
    if n == 2 or n == 3:
        return True
    elif n <= 1 or n % 2 == 0 or n % 3 == 0 :
        return False
    
    for i in range(5, int(math.sqrt(n))+1, 6):
        if n % i == 0 or n % (i+2) == 0:
            return False
        
    return True
```

```python
print(isPrime(29))
print(isPrime(20))
```

# Check if a number is a power of another number

```python
# for y to be the power of x, it must satisfy => y = x^n
# so log(y) / log(x) = n where n is an integer
```

```python
import math

def isPower(x,y):
    ans = math.log(y) / math.log(x)
    
    return ans == math.floor(ans)

print(isPower(2,128))
print(isPower(10,1))
print(isPower(10,1001))
```

# Distance between two points

```python
import math

# distance formula 
# dist. = sqrt((x2-x1)^2 + (y2-y1)^2)

def distanceBtwTwoPoints(x1,y1,x2,y2):
    dist = math.sqrt(math.pow(x2-x1,2) + math.pow(y2-y1,2))
    return dist

print(f"Distance between points is {distanceBtwTwoPoints(3,4,4,3)}")
```

# Valid triangle

--> A triangle is valid if Sum of any two sides is greater than the third side


```python
def isValidTriangle(side1,side2,side3):
    
    a,b,c = side1, side2, side3
    
    if (a+b <= c) or (a+c <= b) or (b+c <= a):
        return 'Invalid'
    else:
        return 'Valid'
    
a, b, c = 7, 10, 5
    
print(isValidTriangle(a,b,c))
    
```

# Count number of times a substring appears in a string

```python
def count_substring(string, sub_string):
    n = len(string)
    count = 0
    
    for i in range(n):
        if string[i] == sub_string[0]:
            if i + len(sub_string) <= n :
                
                k = 1
                for j in range(i+1, i+len(sub_string)):
                    if string[j] != sub_string[k]:
                        break
                    k += 1
                else:
                    count += 1
        
    return count

string = 'ABCDCDC'
sub_string = 'CDC'
# string = 'TestCaseTestCase'
# sub_string = 'CaseT'

count_substring(string, sub_string)
```

# Swap case of string characters

```python
def swap_case(s):
    result = ''
    for c in s:
        if  c.isalpha():
            if c.islower():
                c = c.upper()
            else:
                c = c.lower()
        result += c
    return result

s = 'HackerRank.com presents "Pythonist 2".'
swap_case(s)
```

# Score of runner up

```python
def runnerup(n, arr):
    
    arr = map(int, arr.split())
    arr = list(arr)    
    max_score = float('-inf')
    second_max_score = float('-inf')
    
    for i in range(len(arr)):
        if arr[i] > max_score:
            second_max_score = max_score
            max_score = arr[i]
        elif  second_max_score < arr[i] < max_score:
            second_max_score = arr[i]

    print(second_max_score)
    
n = 5
arr = '2 3 6 6 5'
runnerup(n,arr)
```
