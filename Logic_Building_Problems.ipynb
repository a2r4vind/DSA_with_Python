# 1) Check Even or Odd

```python
# Bitwise AND Operation of Number X with 1 will give 1 if number is Odd else will give 0 if numer is 
```

```python
def isEven(n):
    if (n & 1) == 0:
        return True
    else:
        return False
```

```python
n = 13
if isEven(n):
    print("True")
else:
    print("False")
```

# Multiplication Table


```python
# Recursive approach

def printTable(n, i=1):
    
    if (i == 11):
        return 
    print(f"{n} x {i} = {n*i}")
    i += 1
    printTable(n,i)
```

```python
printTable(5)
```

# Sum of n Natural Numbers

```python
# Naive approach

def findSum(n):
    sum = 0
    i = 1
    while i <= n:
        sum += i
        i += 1
    return sum
```

```python
n = 5
print(findSum(n))
```

```python
# Recursive approach

def findTotal(n):
    
    if n == 1:
        return 1
    return n + findTotal(n - 1)
```

```python
n = 5
print(findTotal(n))
```

```python
# Expected Approach is using the formula

def findNSum(n):
    
    return n * (n + 1) // 2
```

```python
n = 5 
print(findNSum(n))
```

# Sum of first n Squared numbers


```python
def summation(n):
    sq_sum = (n * (n + 1) * (2 * n + 1)) // 6
    return sq_sum
```

```python
n = 10
print(summation(n))
```

# Swap two numbers

```python
a = 3
b = 8

print(f"Before swapping a and b is {a} and {b} respectively.")

a, b = b, a # inbuilt swapping operation

print(f"After swapping : ")
print("a", a)
print("b",b)
```

```python
# Swapping using temporary variable

a = 3
b = 8

print(f"Before swapping a and b is {a} and {b} respectively.")

temp = a
a = b
b = temp

print(f"After swapping : ")
print("a", a)
print("b",b)
```

```python
# swapping using arithmetic operators

a = 3
b = 8

print(f"Before swapping a and b is {a} and {b} respectively.")

a = a + b
b = a - b # b <= (a + b) - b therefore b = a
a = a - b # a <= (a + b) - (a) therefore a = b

print(f"After swapping : ")
print("a", a)
print("b",b)
```

# Find Closest to n and Divisible by m

```python
def closestNum(n,m):
    rem = abs(n) % m
    result = n - rem
    return result

ip1 = 13
m1 = 4
print(closestNum(ip1,m1))
# output => 12

ip2 = -15
m2 = 6
print(closestNum(ip2, m2))
# output => -18
```
