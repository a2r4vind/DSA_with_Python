# 10. Write a program to read seconds and convert them into hours, minutes and seconds.

```python
seconds = int(input("Enter seconds to convert them to Hour: Minutes: Seconds format: "))
hours = seconds // (60 * 60)
minutes = (seconds - hours * 60 * 60 ) // 60
secs = seconds - hours * 60 * 60 - minutes * 60 
print(f"{seconds} in Hours: Minutes: seconds = {hours}: {minutes}: {secs}")
```

# 15. Write a program to read a year and check whether it is a leap year or not.

```python
# logic => year % 4 == 0 and  year % 100 != 0 OR year % 400 == 0 is a leap year else not a leap year
year = int(input("Enter the year to check whether it is a leap year or not: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is NOT a leap year.")
```

# 17. Write a program to read a character and check whether it is an alphabet, digit or special symbol.

```python
char = input('Enter the character to check whether it is an alphabet/digit/special symbol: ')
if len(char) == 1:
    if char.isalpha():
        print(f"Character '{char}' is an alphabet.")
    elif char.isdigit():
        print(f"Character '{char}' is a digit.")
    else:
        print(f"Character '{char}' is a special symbol.")
else:
    print("Please enter exactly one character.")
```
