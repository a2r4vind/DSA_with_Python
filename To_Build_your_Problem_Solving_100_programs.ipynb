# 35. Write a program to find the sum of all digits of a number n.

```python
def digits_sum(n: int) -> int:
    n = abs(n)

    total = 0
    while n > 0:
        digit = n % 10 # last digit is fetched first
        total += digit
        n //= 10
    return total

num = -1234
print(f"Sum of all digits of {num}: {digits_sum(num)}")
```

# 36. Write a program to find the product of all digits of a number n.

```python
def product_digits(n: int) -> int:
    n = abs(n)

    num_of_digits = len(str(n))
    product = 1
    while num_of_digits > 0:
        digit = n // (10 ** (num_of_digits - 1))
        product *= digit
        n %= (10 ** (num_of_digits - 1))
        num_of_digits -= 1
    return product
    
num = -1234
print(f"Product of all digits of {num} = {product_digits(num)}")
```

```python
def count_even_odd_digits(n: int) -> tuple:
    n = abs(n)
    even_count, odd_count = 0, 0
    while n > 0:
        digit = n % 10
        if digit % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
        n //= 10
    return (even_count, odd_count)
num = -123456789
counts = count_even_odd_digits(num)
print(f"Count of even digits is {counts[0]} and odd digits is {counts[1]} in number {num}.")
```

# 41. Write a program to check whether a number n is a palindrome (reads the same reversed).

```python
def check_palindrome_number(n: int) -> bool:
    temp = n
    rev_num = 0
    while temp > 0:
        digit = temp % 10
        rev_num = rev_num * 10 + digit
        temp //= 10
    return True if rev_num == n else False

num = 1234321
print(f"Is {num} is a Palindrome number? {check_palindrome_number(num)}")
```

# 42. Write a program to replace all zeros in a number n with the digit 5.

```python
def replace_zero_with_5(n: int) -> int:
    new_num = 0
    num_of_digits = len(str(n))
    
    while num_of_digits > 0:
        
        # get the first digit
        digit = n // (10 ** (num_of_digits - 1))
        
        # get the remainder ie. number without the first digit
        n %= (10 ** (num_of_digits - 1))

        # replace digit 0 with digit 5
        if digit == 0:
            digit = 5
            
        new_num = new_num * 10 + digit
        
        num_of_digits -= 1
    return new_num

num = 2040608
print(f"Earlier number = {num}")
print(f"After replacing new number = {replace_zero_with_5(num)}")
```
