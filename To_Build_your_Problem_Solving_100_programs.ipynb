# 4. Write a program to read the radius of a circle and print its area and circumference.

```python
from math import pi

PI = float(f"{pi:.4f}") # This takes only the first 4 values after the decimal point in the value of pi
radius = float(input("Enter the radius of a circle: "))
print(f"""Area of Circle = {PI * ( radius ** 2)} sq. units
Circumference of Circle = {2 * PI * radius} units""")
```

# 6. Write a program to swap two numbers using a third variable.

```python
num1 = int(input("Enter the first number : "))
num2 = int(input("Enter the second number: "))
print(f"Numbers before Swapping: {num1} & {num2}")
temp = num2
num2 = num1
num1 = temp
print(f"Numbers after Swapping: {num1} & {num2}")
```

# 7. Write a program to swap two numbers without using a third variable.

```python
# n1 = 6
# n2 = 9
# n1 = n1 + n2 = 6 + 9 = 15
# n2 = n1 - n2  =  15 - 9 = 6
# n1 = n1 - n2 = 15 - 6 = 9

num1 = int(input("Enter the first number : "))
num2 = int(input("Enter the second number : "))
print(f"Numbers before swapping: num1={num1} & num2={num2}")
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2
print(f"Numbers after swapping: num1={num1} & num2={num2}")
```

# 8. Write a program to read a temperature in Celsius and convert it to Fahrenheit.

```python
# deg F = deg C * (9/5)  + 32
deg_C = float(input("Enter the temperature in degree Celsius: "))
deg_F = (deg_C * 9) / 5 + 32
print(f"{deg_C} degree Celsius = {deg_F} degree Fahrenheit.")
```

# 9. Write a program to read the marks of 5 subjects and print the total and average.

```python
marks1 = float(input("Enter the marks of Subject 1: "))
marks2 = float(input("Enter the marks of Subject 2: "))
marks3 = float(input("Enter the marks of Subject 3: "))
marks4 = float(input("Enter the marks of Subject 4: "))
marks5 = float(input("Enter the marks of Subject 5: "))
total_marks = marks1 + marks2 + marks3 + marks4 + marks5
avg_marks = total_marks / 5
print(f"Total marks = {total_marks}\nAverage_marks = {avg_marks}")
```

# 10. Write a program to read seconds and convert them into hours, minutes and seconds.

```python
seconds = int(input("Enter seconds to convert them to Hour: Minutes: Seconds format: "))
hours = seconds // (60 * 60)
minutes = (seconds - hours * 60 * 60 ) // 60
secs = seconds - hours * 60 * 60 - minutes * 60 
print(f"{seconds} in Hours: Minutes: seconds = {hours}: {minutes}: {secs}")
```
