```python
from collections import deque
```

```python
q = deque()
```

```python
q.append(10)
```

```python
q.append(20)
```

```python
q.append(30)
```

```python
print(f"Initial queue : {q}")
```

```python
print("Elements dequeued from the queue: ")
print(q.popleft())
print(q.popleft())
print(q.popleft())
```

```python
print("Queue after removing the elements: ",q)
```
