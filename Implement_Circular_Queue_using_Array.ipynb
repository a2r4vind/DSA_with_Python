```python
class MyCircularQ:
    
    def __init__(self, capacity):
        
        self.capacity = capacity
        
        self.arr = [0] * self.capacity
        
        self.size = 0
        
        self.front = self.rear = -1
    
    # empty check
    def isEmpty(self):
        response = False
        if self.size == 0 :
            self.front = self.rear = -1
            response = True
        return response
    
    # length of circular queue
    def __len__(self):
        
        return self.size
    
    # insertion 
    def enqueue(self, data):
        
        if self.size == self.capacity :
            return "Queue is Full."
        
        if self.rear == -1 and self.front == -1:
            self.front = self.rear = 0
            
        self.rear = (self.front + self.size) % self.capacity
        
        self.arr[self.rear] = data
        self.size += 1
    
    # deletion 
    def dequeue(self):
        
        if self.isEmpty():
            return "Circular Queue is empty."
        
        data = self.arr[self.front]
        self.front = (self.front + 1) % self.capacity
        self.size -= 1
        return data
    
    # method to get the element at the front
    def getFront(self):
        
        if self.isEmpty():
            return "Circular Queue is Empty"
        
        return self.arr[self.front]
    
    # method to get the element at the rear
    def getRear(self):
        
        if self.isEmpty():
            return "Circular Queue is Empty"
        
        return self.arr[self.rear]
    
    # method to traverse the Circular Queue
    def printQueue(self):
        
        # empty check
        if self.isEmpty():
            return "Circular Queue is Empty"
        
        # curr pointer to get the current element
        curr = self.front
        
        # length of Circular Queue
        n = self.size
        
        i = 1
        while i <= n:
            print(f"{self.arr[curr]}",end="")
            if i is not n:
                print(" -> ",end="")
            curr = (curr + 1) % self.capacity
            i += 1
        print()
```
