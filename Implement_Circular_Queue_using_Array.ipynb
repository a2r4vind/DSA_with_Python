```python
class MyCircularQ:
    
    def __init__(self, capacity):
        
        self.capacity = capacity
        
        self.arr = [0] * self.capacity
        
        self.size = 0
        
        self.front = self.rear = -1
        
    def isEmpty(self):
        response = False
        if self.size == 0 :
            self.front = self.rear = -1
            response = True
        return response
    
    def __len__(self):
        
        return self.size
        
    def enqueue(self, data):
        
        if self.size == self.capacity :
            return "Queue is Full."
        
        if self.rear == -1 and self.front == -1:
            self.front = self.rear = 0
            
        self.rear = (self.front + self.size) % self.capacity
        
        self.arr[self.rear] = data
        self.size += 1
        
    def dequeue(self):
        
        if self.isEmpty():
            return "Circular Queue is empty."
        
        data = self.arr[self.front]
        self.front = (self.front + 1) % self.capacity
        self.size -= 1
        return data
        
#     def printQueue(self):
        
#         if self.isEmpty():
#             return "Circular Queue is Empty"
        
#         curr = self.front
        
#         last = self.rear
        
#         while curr is not last:
#             print(f"{self.arr[curr]}",end="")
#             if curr is not last:
#                 print(" -> ",end="")
#             curr = (curr + 1) % self.size
            
#         print(self.arr[curr])
            
#         print()
```
